## Perform the following steps:

1. Construct a tensor of rank 2 (matrix) with the following dimension:
    1. Number os rows = 2
    1. Number of columns = 3

1. Apply Softmax on each column and row, respectively.

```python
import torch 

def sum_over_softmax_values(arr, dim):
    print(f'dim = {dim}')
    
    softmax = torch.softmax(arr, dim = dim)
    print(softmax)
    print(softmax.shape)
    
    summation = torch.sum(softmax, dim = dim)
    print(summation)
    print(summation.shape)

tensor_with_random_values = torch.randn(2, 3)
print(tensor_with_random_values)    
print(tensor_with_random_values.shape)    

print()
print()
print()

sum_over_softmax_values(tensor_with_random_values, 0) 
sum_over_softmax_values(tensor_with_random_values, -2)

print()
print()
print()

sum_over_softmax_values(tensor_with_random_values, 1)  
sum_over_softmax_values(tensor_with_random_values, -1) 
```

## Perform the following steps:

1. Construct a tensor of rank 3 (cube) with the following dimension:
    1. Number os rows = 9
    1. Number of columns = 12
    1. Depth size = 6

1. Apply Softmax on each (row, column), (depth, column), and (depth, row), respectively.

```python
import torch 

def sum_over_softmax_values(arr, dim):
    print(f'dim = {dim}')
    
    softmax = torch.softmax(arr, dim = dim)
    print(softmax.shape)
    
    summation = torch.sum(softmax, dim = dim)
    print(summation)
    print(summation.shape)
    
tensor_with_random_values = torch.randn(6, 9, 12)
print(tensor_with_random_values.shape)    
print()

sum_over_softmax_values(tensor_with_random_values, -3) 
sum_over_softmax_values(tensor_with_random_values, 0)

print()
print()
print()

sum_over_softmax_values(tensor_with_random_values, -2)    
sum_over_softmax_values(tensor_with_random_values, 1)

print()
print()
print()

sum_over_softmax_values(tensor_with_random_values, -1)      
sum_over_softmax_values(tensor_with_random_values, 2)    
```
